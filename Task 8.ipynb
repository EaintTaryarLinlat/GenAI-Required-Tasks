Block highlights Bitcoin custody mechanics (irreversible loss/compromise of private keys; hot-wallet and third-party custodian exposure) and links them to litigation and regulatory enforcement risk.  It flags evolving crypto regulation, including new state licensing (e.g., Louisiana effective Jan 1, 2023) and admits that, because of blockchain’s design, it may be “technically infeasible” to prevent all prohibited transactions—an AML/sanctions compliance friction point.  Block also signals elevated supervisory pressure via ongoing audits, inquiries, and investigations, implying recurring remediation cost and headline risk.  It further discloses operational resiliency gaps (systems not fully redundant; disaster recovery not sufficient for all eventualities).

PayPal emphasizes efficiency through modernization and resiliency efforts, while warning that costly re-platforming and persistent interruptions could trigger regulatory scrutiny and even loss of licenses.  PayPal is also more explicit about AI/ML and privacy as governance risks, linking evolving privacy obligations to AI/ML and crypto/blockchain technologies, and noting enforcement hooks such as regulators considering its compliance history.

Overall, Block’s AI & Bitcoin push expands operational/legal risk surface area versus PayPal’s conservative posture; Bitcoin risks are disclosed clearly, but AI/model-governance risks appear comparatively less explicit and should be treated as a latent exposure.
